```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
import xgboost as xgb
from xgboost import plot_importance, plot_tree
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
filename = 'PJMW_hourly.csv'
path = Path(r"C:\Users\Matth\git\DataAnalysisWorkbooks\XGBoost\TimeSeries_EnergyUsage")

data = pd.read_csv(path/filename, index_col=[0], parse_dates=[0])
```

```python
color_pal = ["#F8766D", "#93AA00", "#D39200", "#00BA38", "#00C19F", "#00B9E3", "#619CFF", "#DB72FB"]
_ = data.plot(style='.', figsize=(15,5), color=color_pal[0], title='PJM West')
```

```python
# Split the data into a training and testing dataset. (Test set = after Jan 1 2015)
split_date = '01-Jan-2015'
data_train = data.loc[data.index <= split_date].copy()
data_test = data.loc[data.index > split_date].copy()
```

```python
_ = data_test \
    .rename(columns = {'PJME_MW' : 'TEST_SET'}) \
    .join(data_train.rename(columns={'PJMW_MW': 'TRAINING SET'}), how='outer') \
    .plot(figsize=(15,5), title='PJM West', style='.')
```

```python
# Create the time series features
```

```python
def create_features(df, label=None):
    """
    Creates time series features from datetime index
    """
    df['date'] = df.index
    df['hour'] = df['date'].dt.hour
    df['dayofweek'] = df['date'].dt.dayofweek
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['dayofmonth'] = df['date'].dt.day
    df['weekofyear'] = df['date'].dt.weekofyear
    
    X = df[['hour','dayofweek','quarter','month','year',
           'dayofyear','dayofmonth','weekofyear']]
    if label:
        y = df[label]
        return X, y
    return X
```

```python
X_train, y_train = create_features(data_train, label='PJMW_MW')
X_test, y_test = create_features(data_test, label='PJMW_MW')
```

```python
reg = xgb.XGBRegressor(n_estimators=1000)
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        early_stopping_rounds=50,
       verbose=True) # Change verbose to True if you want to see it train
```

"Feature importance" is a great way to get a general idea about which features the model is relying on most to make the prediction. This is a metric that simply sums up how many times each feature is split on.

We can see that the day of year was most commonly used to split trees, while hour and year came in next. Quarter has low importance due to the fact that it could be created by different dayofyear splits.

```python
_ = plot_importance(reg, height=0.9)
```

Now run the model on the validation set.

```python
data_test['MW_Prediction'] = reg.predict(X_test)
data_all = pd.concat([data_test, data_train], sort=False)
```

```python
_ = data_all[['PJMW_MW','MW_Prediction']].plot(figsize=(15, 5))
```

Error measurements:  
RMSE - Mean square error  
MAE - Mean average error  
MAPE - Mean average percent error

```python
mean_squared_error(y_true=data_test['PJMW_MW'],
                   y_pred=data_test['MW_Prediction'])
```

```python
mean_absolute_error(y_true=data_test['PJMW_MW'],
                   y_pred=data_test['MW_Prediction'])
```

```python
# Note MAPE is not a function in scikit-learn

def mean_absolute_percentage_error(y_true, y_pred): 
    """Calculates MAPE given y_true and y_pred"""
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mean_absolute_percentage_error(y_true=data_test['PJMW_MW'],
                   y_pred=data_test['MW_Prediction'])
```
