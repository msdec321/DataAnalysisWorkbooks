```python
import pandas as pd
```

```python
saleDates_data = pd.read_csv('order_dates.csv', index_col=[0], parse_dates=[0])
```

```python
color_pal = ["#F8766D", "#93AA00", "#D39200", "#00BA38", "#00C19F", "#00B9E3", "#619CFF", "#DB72FB"]
_ = saleDates_data.plot(style='.', figsize=(15,5), color=color_pal[0], title='Daily sales')
```

```python
_ = priceQuantity_data.plot(x='product_id', y='quantity', style='o', figsize=(15,5), color=color_pal[0], title='Quantity by id')
```

```python
_ = priceQuantity_data.plot(x='product_id', y='list_price', figsize=(15,5), color=color_pal[0], title='Price by id')
```
