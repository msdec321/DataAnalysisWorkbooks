```python
months_list = ['2009-01', '2009-03', '2009-05', '2009-07', '2009-09', '2009-011',
               '2010-01', '2010-03', '2010-05', '2010-07', '2010-09', '2010-011',
               '2011-01', '2011-03', '2011-05', '2011-07', '2011-09', '2011-011',
               '2012-01', '2012-03', '2012-05', '2012-07', '2012-09', '2012-011',
               '2013-01', '2013-03', '2013-05', '2013-07', '2013-09', '2013-011',
               '2014-01', '2014-03', '2014-05', '2014-07', '2014-09', '2014-011',
               '2015-01', '2015-03', '2015-05', '2015-07', '2015-09', '2015-011',
               '2016-01', '2016-03', '2016-05', '2016-07', '2016-09', '2016-011',
               '2017-01', '2017-03', '2017-05', '2017-07', '2017-09', '2017-011']  # Date format

months_labels = ['2009', '', '', '', '', '', 
                 '2010', '', '', '', '', '',
                 '2011', '', '', '', '', '',
                 '2012', '', '', '', '', '',
                 '2013', '', '', '', '', '',
                 '2014', '', '', '', '', '',
                 '2015', '', '', '', '', '',
                 '2016', '', '', '', '', '',
                 '2017', '', '', '', '', '']  # xtick labels for matplotlib

energyDictionary = {
  "avg_power": { "2009" : [0., 0., 0., 0., 0., 0.],
                 "2010" : [0., 0., 0., 0., 0., 0.],
                 "2011" : [0., 0., 0., 0., 0., 0.],
                 "2012" : [0., 0., 0., 0., 0., 0.],
                 "2013" : [0., 0., 0., 0., 0., 0.],
                 "2014" : [0., 0., 0., 0., 0., 0.],
                 "2015" : [0., 0., 0., 0., 0., 0.],
                 "2016" : [0., 0., 0., 0., 0., 0.],
                 "2017" : [0., 0., 0., 0., 0., 0.] } ,
    
      "power": { "2009" : [0., 0., 0., 0., 0., 0.],
                 "2010" : [0., 0., 0., 0., 0., 0.],
                 "2011" : [0., 0., 0., 0., 0., 0.],
                 "2012" : [0., 0., 0., 0., 0., 0.],
                 "2013" : [0., 0., 0., 0., 0., 0.],
                 "2014" : [0., 0., 0., 0., 0., 0.],
                 "2015" : [0., 0., 0., 0., 0., 0.],
                 "2016" : [0., 0., 0., 0., 0., 0.],
                 "2017" : [0., 0., 0., 0., 0., 0.] } ,
      
      "price": { "2009" : [0., 0., 0., 0., 0., 0.],
                 "2010" : [0., 0., 0., 0., 0., 0.],
                 "2011" : [0., 0., 0., 0., 0., 0.],
                 "2012" : [0., 0., 0., 0., 0., 0.],
                 "2013" : [0., 0., 0., 0., 0., 0.],
                 "2014" : [0., 0., 0., 0., 0., 0.],
                 "2015" : [0., 0., 0., 0., 0., 0.],
                 "2016" : [0., 0., 0., 0., 0., 0.],
                 "2017" : [0., 0., 0., 0., 0., 0.]  } ,

    "revenue": { "2009" : [0., 0., 0., 0., 0., 0.],
                 "2010" : [0., 0., 0., 0., 0., 0.],
                 "2011" : [0., 0., 0., 0., 0., 0.],
                 "2012" : [0., 0., 0., 0., 0., 0.],
                 "2013" : [0., 0., 0., 0., 0., 0.],
                 "2014" : [0., 0., 0., 0., 0., 0.],
                 "2015" : [0., 0., 0., 0., 0., 0.],
                 "2016" : [0., 0., 0., 0., 0., 0.],
                 "2017" : [0., 0., 0., 0., 0., 0.]  } }
```

```python
temp_year, temp_month = ['', ''], ['', '']
temp_month_index = [0, 0]
total_power, total_price, total_revenue, count = 0., 0., 0., 0

for i, price in enumerate(energyList, 1):
       
    try:
        temp_year[0] = temp_year[1]
        temp_year[1] = energyList[i][1]
        
        temp_month[0] = temp_month[1] 
        if energyList[i][2] in ['1', '2']:
            temp_month[1] = 'Jan'
            temp_month_index[0] = temp_month_index[1]
            temp_month_index[1] = 0
        
        elif energyList[i][2] in ['3', '4']: 
            temp_month[1] = 'Mar'
            temp_month_index[0] = temp_month_index[1]
            temp_month_index[1] = 1

        elif energyList[i][2] in ['5', '6']:
            temp_month[1] = 'May'
            temp_month_index[0] = temp_month_index[1]
            temp_month_index[1] = 2
            
        elif energyList[i][2] in ['7', '8']: 
            temp_month[1] = 'Jul'
            temp_month_index[0] = temp_month_index[1]
            temp_month_index[1] = 3

        elif energyList[i][2] in ['9', '10']:
            temp_month[1] = 'Sep'
            temp_month_index[0] = temp_month_index[1]
            temp_month_index[1] = 4
            
        elif energyList[i][2] in ['11', '12']:
            temp_month[1] = 'Nov'
            temp_month_index[0] = temp_month_index[1]
            temp_month_index[1] = 5

        if i == 1: continue  # Skip first row of data

        if temp_month[0] == temp_month[1]:
            total_price += float(energyList[i][6])
            total_power += float(energyList[i][5])
            total_revenue += float(energyList[i][6]) * float(energyList[i][5])
            count += 1

        else:
            energyDictionary["avg_power"][temp_year[0]][temp_month_index[0]] = total_power / count
            energyDictionary["power"][temp_year[0]][temp_month_index[0]] = total_power
            energyDictionary["price"][temp_year[0]][temp_month_index[0]] = total_price / count
            energyDictionary["revenue"][temp_year[0]][temp_month_index[0]] = total_revenue
            total_power = 0.
            total_price = 0.
            total_revenue = 0.
            count = 0
            
    except IndexError: pass
    
energyDictionary["avg_power"]['2017'][5] = total_power / count  # Fill average for Nov 2017
energyDictionary["power"]['2017'][5] = total_power
energyDictionary["price"]['2017'][5] = total_price / count
energyDictionary["revenue"]['2017'][5] = total_revenue

avg_powerByMonth, powerByMonth, priceByMonth, revenueByMonth = [], [], [], []
for month in energyDictionary["power"]:
    for i, item in enumerate(energyDictionary["avg_power"][month]):
        avg_powerByMonth.append(energyDictionary["avg_power"][month][i])
        powerByMonth.append(energyDictionary["power"][month][i])
        priceByMonth.append(energyDictionary["price"][month][i])
        revenueByMonth.append(energyDictionary["revenue"][month][i])
```
