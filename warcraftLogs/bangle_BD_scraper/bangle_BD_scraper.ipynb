## This notebook scrapes bangle/blue dragon data from speedrun logs
## Multiprocessing is not supported

## Import packages

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import shutil, time, os
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

from importlib import reload

import src
reload(src)
```

## Configurations

```python
# Set path variables
path_to_ublock = r'C:\Users\Matth\Documents\1.42.4_6'

# Specify which rank to scrape up to
nParses = 1030
```

## Navigate to WCL

```python
speedclear_url, browser = src.load_scraper(path_to_ublock, chrome=True, firefox=False)
```

## Main code loop

```python
speedclear_url = f'https://classic.warcraftlogs.com/zone/rankings/1011/#metric=speed&boss=704'
browser.get(speedclear_url)
time.sleep(2)

# Update recorded ranks in spreadsheet
already_recorded_indices = src.get_latest_ranks(browser, nParses)

page = 1

for i in range(1, nParses):
    
    # Start scraping after the last recorded rank
    if i <= already_recorded_indices[-1]: 
        if i % 50 == 0: 
            page += 1
            speedclear_url = f'https://classic.warcraftlogs.com/zone/rankings/1011/#metric=speed&boss=704&page={page}'
            browser.get(speedclear_url)
            time.sleep(3) 
        continue

    row = browser.find_element_by_id(f"row-704-{i}")

    rank = row.text.split("\n")[0]
    guild_name = row.text.split("\n")[1]
    duration = row.text.split("\n")[3].split(" ")[-1]
    
    log = browser.find_element_by_link_text(guild_name)
    log.click()
    time.sleep(2)

    src.click_on_element_by_id(browser, "filter-summary-tab")  
    time.sleep(3)

    summary_rows = browser.find_elements_by_class_name("composition-row")
    healers = summary_rows[2]

    td = healers.find_elements_by_css_selector("td")
    healer = td[1].find_elements_by_class_name("composition-entry")

    # Check if Druid healer in log
    isDruid = False
    for item in healer:
        temp = item.find_element_by_css_selector("a")
        if temp.get_attribute("class") == "Druid":
            name = temp.text
            isDruid = True   
            
    if isDruid:
        print(f"Rank: {rank}, Guild name: {guild_name}. Druid Healer in this raid! Name is: {name}")
        
    else:
        print(f"Rank: {rank}, Guild name: {guild_name}. No Druid Healer in this raid.")
        print("------")
        
        if i % 50 == 0: 
            page += 1
            speedclear_url = f'https://classic.warcraftlogs.com/zone/rankings/1011/#metric=speed&boss=704&page={page}'
            browser.get(speedclear_url)
            time.sleep(2)
        
        continue
        
    summary_page = browser.current_url
    boss_time_menu = browser.find_element_by_id("filter-fight-boss-text")
    boss_time_menu.click()
    time.sleep(1)

    time_in_combat = src.get_time_in_combat(browser)
    browser.get(summary_page)
    time.sleep(2)
        
    player = browser.find_element_by_link_text(name)
    player.click()
    time.sleep(2)
        
    has_bd, has_bangle = False, False

    equipment = browser.find_elements_by_class_name("epic")

    # Check if Blue Dragon and Bangle in equipment list.
    for item in equipment:
        if item.text in ["Darkmoon Card: Blue Dragon", "暗月卡片：蓝龙"]: 
            has_bd = True

    equipment = browser.find_elements_by_class_name("rare")

    for item in equipment:
        if item.text in ["Bangle of Endless Blessings", "无尽祝福法链"]: 
            has_bangle = True

    print(f"Blue Dragon: {has_bd}, Bangle: {has_bangle}")
    
    if has_bd and has_bangle:
        
        player_df = pd.DataFrame(pd.np.empty((0, 25)))
        player_df.columns = ["Rank", "Druid", "Guild", "Duration", "Time in combat (s)", "Avg Int", "Avg Spirit", "BD uptime (s)", "Bangle use uptime (s)", "# BD procs", "# Bangle uses", "Total overlap time (s)", "# bangle overlaps", "# bangle non-overlaps", "Avg time between BD procs (s)", "Avg time between bangle ready and BD proc", "BD mana received", "Bangle mana received", "BD+Bangle mana received", "BD+Bangle non overlap mana", "BD mp5", "Bangle use mp5", "BD+Bangle mp5", "Bangle mp5 normalized", "BD+Bangle mp5 normalized"]

        # Get average primary stats
        intellect, spirit = src.get_int_spirit(browser)
        
        # Navigate to Aura of Blue Dragon proc tables
        buffs_tab = browser.find_element_by_id("filter-buffs-tab")
        buffs_tab.click()
        time.sleep(2)
        
        buff_url = browser.current_url
        
        buff_list = browser.find_element_by_id("filter-ability-text")
        buff_list.click()
        time.sleep(1)
        
        # Currently supports English and Chinese logs
        try:
            bd_link = browser.find_element_by_link_text("蓝龙光环")
        except:
            bd_link = browser.find_element_by_link_text("Aura of the Blue Dragon")
            
        bd_link.click()
        time.sleep(1)
        
        body = browser.find_element_by_css_selector('body')
        body.click()
        body.send_keys(Keys.PAGE_UP) 
        time.sleep(1)
        
        src.click_on_element_by_id(browser, "filter-events-tab")    
        time.sleep(1)
        
        # Get a list of timestamps for Blue Dragon procs, uptime, avg time between procs
        all_bd_procs = src.get_bd_list(browser)
        time_between_BD = src.get_seconds_between_BD_procs(all_bd_procs)
        bd_uptime, bd_procs = src.get_BD_procs(all_bd_procs)
        
        browser.get(buff_url)
        time.sleep(2)
        
        # Navigate to Endless Blessings proc tables
        buff_list = browser.find_element_by_id("filter-ability-text")
        buff_list.click()
        time.sleep(1)
        
        # Currently supports English and Chinese logs
        try:
            bangle_link = browser.find_element_by_link_text("无尽祝福")
        except:
            bangle_link = browser.find_element_by_link_text("Endless Blessings")
            
        bangle_link.click()
        time.sleep(1)
        
        body.send_keys(Keys.PAGE_UP) 
        time.sleep(1)
        
        src.click_on_element_by_id(browser, "filter-events-tab")    
        time.sleep(1)
        
        # Get list of bangle use timestamps, uptime, overlap, etc
        all_bangle_uses = src.get_bangle_list(browser)
        bangle_use_uptime, bangle_uses = src.get_bangle_uses(all_bangle_uses)
        overlap_uptime, overlap_occurances = src.get_total_overlap_time(all_bd_procs, all_bangle_uses)
        
        # Get aura uptimes, total mana received from each, and calculate mp5 for each
        pure_bd_uptime = bd_uptime - overlap_uptime
        pure_bangle_use_uptime = bangle_use_uptime - overlap_uptime
        bd_mana, bangle_mana, bd_bangle_mana, bd_bangle_nonoverlap_mana = src.calc_mana(intellect, spirit, pure_bd_uptime, overlap_occurances, pure_bangle_use_uptime, overlap_uptime)
        bd_mp5, bangle_mp5, bd_bangle_mp5, bangle_mp5_normalized, bd_bangle_mp5_normalized = src.calc_mp5(time_in_combat, overlap_uptime, time_between_BD, bangle_uses, overlap_occurances, bd_mana, bangle_mana, bd_bangle_mana, bd_bangle_nonoverlap_mana)
        
        # Calculate average time between bangle cooldown ready and next blue dragon proc
        bd_aura_clock = src.get_bd_proc_times(all_bd_procs)
        bangle_ready_times = src.get_bangle_ready_times(all_bangle_uses)
        time_between_bangle_bd = src.get_time_between_bangle_use_and_bdproc(bangle_ready_times, bd_aura_clock)
        
        print(f"BD uptime (seconds): {bd_uptime}, BD procs: {bd_procs}")
        print(f"Bangle use uptime (seconds): {bangle_use_uptime}, Bangle uses: {bangle_uses}")
        print(f"Total BD+bangle overlap time (seconds): {overlap_uptime}")
        print(f"Times bangle used with BD: {overlap_occurances}, Times bangle used without BD: {bangle_uses - overlap_occurances}")
           
        # Export to spreadsheet
        to_append = [str(rank), name, guild_name, duration, time_in_combat, intellect, spirit, bd_uptime, bangle_use_uptime, str(bd_procs), str(bangle_uses), overlap_uptime, str(overlap_occurances), str(bangle_uses - overlap_occurances), time_between_BD, time_between_bangle_bd, bd_mana, bangle_mana, bd_bangle_mana, bd_bangle_nonoverlap_mana, bd_mp5, bangle_mp5, bd_bangle_mp5, bangle_mp5_normalized, bd_bangle_mp5_normalized]
        src.export_to_csv(to_append, player_df, name)
        os.remove(f"bangleBD_{name}.csv")

    print("------")
    
    browser.get(speedclear_url)
    time.sleep(2)
    
    if i % 50 == 0: 
        page += 1
        speedclear_url = f'https://classic.warcraftlogs.com/zone/rankings/1011/#metric=speed&boss=704&page={page}'
        browser.get(speedclear_url)
        time.sleep(2)
    
```
